```python
years = df.columns[4:]
values = df[years].sum()
plt.figure (figsize=(30,20))
plt.barh(years,values,edgecolor='red')
plt.xlabel('Total Values')
plt.ylabel('Year', size=20)
plt.title('Total Values per Year', size=20)
plt.show()
```

```python
for i in years:
  fig = plt.figure(figsize=(5,5))
  plt.hist(df[i],color ='green',bins=10)
  plt.xlabel(i)
  plt.show()
```
